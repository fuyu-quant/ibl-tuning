```python
import torch
from peft import PeftModel
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from datasets import load_dataset

import re

model_name = 'meta-llama/Llama-2-13b-hf'
#dataset_name = 'ibl-regression-ver1-linear'
#dataset_name = 'ibl-regression-ver1-branch'
#dataset_name = 'ibl-regression-ver1-mix'
#dataset_name = 'ibl-regression-ver1-all'
#dataset_name = 'ibl-regression-ver2-linear'
dataset_name = 'ibl-regression-ver2-branch_easy'
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_use_double_quant=True,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=torch.bfloat16
    ),
    device_map={"":0}
)
```

```python
output_path = f"/content/drive/My Drive/ibl/{model_name.replace('/', '_')}_{dataset_name}"

model = PeftModel.from_pretrained(
    model,
    output_path + "/checkpoint-1000/adapter_model/",
    device_map={"":0}
)
model.eval()
```

```python
login(token=userdata.get('huggingface-write'))
model.push_to_hub(f"fuyu-quant/{dataset_name}")
```

```python
def ibl_test(input, true_data):
    prompt = f"### Instruction: {input}\n\n### Response: "

    # 推論の実行
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda:0")
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=300
            )
    output = tokenizer.decode(outputs[0], skip_special_tokens=True)
    output = output.split("### Response: ")[-1]
    print('出力データ:\n', output)
    print('------')
    print('正解データ:\n', true_data)
    return
```

## Train data

```python
train_dataset = load_dataset(f"fuyu-quant/{dataset_name}", split='train')
```

```python
i = 1

input_data = train_dataset[i]['instruction']
output_data = train_dataset[i]['output']
ibl_test(input_data, output_data)
```

## Test data

```python
test_dataset = load_dataset(f"fuyu-quant/{dataset_name}", split='test')
```

```python
i = 3

input_data = test_dataset[i]['instruction']
output_data = test_dataset[i]['output']
ibl_test(input_data, output_data)
```
